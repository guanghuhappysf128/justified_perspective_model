```python
import pymongo
import os
import json

my_client = pymongo.MongoClient("mongodb://localhost:27017",username="admin",password="90054")
my_db = my_client['new_final_result']


all_instance_collection = my_db['all_instance']
job_record_backup_collection = my_db['job_record_backup']
job_record_collection = my_db['job_record']

# my_collection = my_db['job_index']

# job_record_collection.update_many({},{ "$rename": { "search_name": "search"}})
# job_record_collection.update_many({},{ "$rename": { "num_of_agent": "num_of_agents"}})


search_name = "ngreedy"

# with open('/home/ghu2004/Downloads/001_job_record_backup.json','r') as f:
existing_dict = dict()
query = {"search":search_name}


for item in all_instance_collection.find(query):

    number_of_agents = item['num_of_agents']
    if not number_of_agents in existing_dict.keys():
        existing_dict[number_of_agents] = dict()
    
    designed_goal_depth = item['designed_goal_depth']
    if not designed_goal_depth in existing_dict[number_of_agents].keys():
        existing_dict[number_of_agents][designed_goal_depth] = dict()

    goal_size = item['total_goal_size']
    if not goal_size in existing_dict[number_of_agents][designed_goal_depth].keys():
        existing_dict[number_of_agents][designed_goal_depth][goal_size] = dict()

    init_name = item['init_name']
    if not init_name in existing_dict[number_of_agents][designed_goal_depth][goal_size].keys():
        existing_dict[number_of_agents][designed_goal_depth][goal_size][init_name] = list()

    goal_index = item['goal_index']
    if not goal_index in existing_dict[number_of_agents][designed_goal_depth][goal_size][init_name]:
        existing_dict[number_of_agents][designed_goal_depth][goal_size][init_name].append(goal_index)
    else:
        print("there are something wrong, it should not be the same")
        print(item)


job_record_list = list()
for item in job_record_backup_collection.find(query):
    designed_goal_depth = item['designed_goal_depth']
    number_of_agents = item['num_of_agents']
    init_name = item['init_name']
    goal_size = item['goal_size']
    goal_index = item['goal_index']
    if number_of_agents in existing_dict.keys():
        if designed_goal_depth in existing_dict[number_of_agents].keys():
            if goal_size in existing_dict[number_of_agents][designed_goal_depth].keys():
                if init_name in existing_dict[number_of_agents][designed_goal_depth][goal_size].keys():
                    if goal_index in existing_dict[number_of_agents][designed_goal_depth][goal_size][init_name]:
                        continue  
    job_record_list.append(item)


print(len(job_record_list))
job_record_collection.insert_many(job_record_list)
```
